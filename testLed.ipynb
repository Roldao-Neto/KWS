#### Test with Python:

```python
from gpiozero import LED
from time import sleep
```

The following code simply blinks the LED connected to the GPIO4 port every second:

```python
led = LED(4)

try:
    while True:
        led.on()
        sleep(1)
        led.off()
        sleep(1)

except KeyboardInterrupt:
    led.close()
```
